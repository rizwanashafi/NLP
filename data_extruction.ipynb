### Let's do some Text preprocessing and cleaning

```python
import re
from nltk.corpus import stopwords
from textblob import TextBlob

# Load NLTK stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def transform(text):

    review = re.sub('[^a-zA-Z0-9]', ' ',text)  # except small and capital letters and numeric remove everythong.
    review = review.lower()                    # lower it.
    review = review.split()

    review = [word for word in review if not word in stopwords.words('english')]   # remove stopwords.
    review = ' '.join(review)
    return review


df['Transform_Text'] = df['Text'].apply(transform)
```

# **text sentiment**

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity, analysis.sentiment.subjectivity
```

# **avg_sentence_length**

```python
def avg_sentence_length(text):
    sentences = re.split(r'[.!?]', text)
    sentence_lengths = [len(sentence.split()) for sentence in sentences if len(sentence.split()) > 0]
    return sum(sentence_lengths) / len(sentence_lengths) if len(sentence_lengths) > 0 else 0
```

# **percentage_complex_words**

```python
def percentage_complex_words(text):
    words = text.split()
    complex_words = [word for word in words if syllable_count(word) >= 3]
    return (len(complex_words) / len(words)) * 100 if len(words) > 0 else 0
```

# **fog_index**

```python
def fog_index(text):
    return 0.4 * (avg_sentence_length(text) + percentage_complex_words(text))
```

# **complex_word_count**

```python
def complex_word_count(text):
    words = text.split()
    complex_words = [word for word in words if syllable_count(word) >= 3]
    return len(complex_words)
```

# **syllable_per_word**

```python
def syllable_per_word(text):
    words = text.split()
    syllable_count_total = sum(syllable_count(word) for word in words)
    return syllable_count_total / len(words) if len(words) > 0 else 0
```

# **personal_pronouns**

```python
def personal_pronouns(text):
    pronouns = ['i', 'me', 'my', 'mine', 'myself', 'you', 'your', 'yours', 'yourself', 'he', 'him', 'his', 'himself',
                'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'we', 'us', 'our', 'ours', 'ourselves', 'they',
                'them', 'their', 'theirs', 'themselves']
    words = text.lower().split()
    return sum(1 for word in words if word in pronouns)
```

# **avg_word_length**

```python
def avg_word_length(text):
    words = text.split()
    word_lengths = [len(word) for word in words]
    return sum(word_lengths) / len(word_lengths) if len(word_lengths) > 0 else 0
```

# **observations:**

> **1. Text:** This column contains the processed text of the articles, where stopwords have been removed, and the text is preprocessed for further analysis.



> **2. URL:** This column contains the original URLs of the articles.

> **3. Positive_Score:** The positivity score of the text, calculated using sentiment analysis. It represents the degree of positive sentiment in the text.

> **4. Negative_Score:** The negativity score of the text, which is complementary to the positive score. It represents the degree of negative sentiment in the text.

> **5. Subjectivity_Score:** The subjectivity score of the text, indicating how subjective or objective the text is.

> **6. Avg_Sentence_Length:** The average length of sentences in the text, measured in the number of words.

> **7. Percentage_Complex_Words:** The percentage of complex words in the text. Complex words are those with three or more syllables.

> **8. Fog_Index:** A readability measure that estimates the years of formal education needed to understand the text. It is calculated based on average sentence length and the percentage of complex words.

> **9. Personal_Pronouns:** The count of personal pronouns (e.g., I, me, you, he, she) in the text.

> These columns provide various metrics and insights into the linguistic and sentiment characteristics of the analyzed text.
